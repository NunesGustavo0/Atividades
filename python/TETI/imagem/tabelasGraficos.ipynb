```python
graf = sns.scatterplot(x='ENGINESIZE', y='CO2EMISSIONS', data = df_selecao, \
                      hue = 'ENGINESIZE')
```

```python
treino1 = df_selecao[:(int(len(df_selecao)*0.7))]
teste1 = df_selecao[(int(len(df_selecao)*0.7)):]
```

```python
len(treino1), len(teste1)
```

```python
modelo_regl = linear_model.LinearRegression()
```

```python
import numpy as np
```

```python
treino_x = np.array(treino1[['ENGINESIZE']])
treino_y = np.array(treino1[['CO2EMISSIONS']])
```

```python
modelo_regl.fit(treino_x, treino_y) ##.fit = treina
```

```python
print('Coeficiente: ', modelo_regl.coef_)
print('Interceptao: ', modelo_regl.intercept_)
```

```python
plt.scatter(treino1['ENGINESIZE'], \
           treino1['CO2EMISSIONS'], color = 'green')
plt.plot(treino_x, modelo_regl.coef_*treino_x + modelo_regl.intercept_, '-r')
plt.xlabel('Engine size')
plt.ylabel('Emisso de co2')
plt.show()
```

```python
def predicao_co2(dados_entrada, intercept, slope):
    pred = dados_entrada* slope+intercept
    return pred
```

```python
potencia = 1.7
emissao_est = predicao_co2(potencia, \
                           modelo_regl.intercept_[0], \
                          modelo_regl.coef_[0][0])
print('Emisso de Co2 para potncia de ', potencia, ' : ', emissao_est)
```

```python
import numpy as np
from sklearn.datasets import load_digits
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as KMN
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
digits = load_digits()
```

```python
len(digits.feature_names)
```

```python
digits.target
```

```python
digits.target_names
```

```python
fig = plt.figure(figsize=(10,8))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1,\
                   hspace=0.05, wspace=0.05)
for i in range(100):
    graf = fig.add_subplot(10, 10, i+1, xticks=[], yticks=[])
    graf.imshow(digits.images[i], cmap = plt.cm.gray, interpolation = 'lanczos')
```

```python
x = digits.data
y = digits.target
```

```python
x.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,\
                                                   random_state=0)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
modelo_kmn = KMN(n_neighbors = 5)
```

```python
modelo_kmn.fit(x_train, y_train)
```

```python
predicoes = modelo_kmn.predict(x_test)
```

```python
predicoes
```

```python
acuracia = modelo_kmn.score(x_test, y_test)
print('Acurcia: ', acuracia)
```

```python
matriz_confusao = metrics.confusion_matrix(y_test, predicoes)
print('Matriz de confuso: ')
print(matriz_confusao)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo_RF = RandomForestClassifier(n_estimators = 750, random_state = 0)
modelo_RF.fit(x_train, y_train)
```

```python
predicoes = modelo_RF.predict(x_test)
```

```python
acuracia = modelo_RF.score(x_test, y_test)
print('Acuracia', acuracia)
```

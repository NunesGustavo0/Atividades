```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_2img(im1, tit1, im2, tit2, tam, ticks = False):
    plt.figure(figsize = tam)
    plt.subplot(1,2,1), plt.imshow(im1), plt.title(tit1)
    if not ticks:
        plt.xticks([]), plt.yticks([])
    plt.subplot(1,2,2), plt.imshow(im2), plt.title(tit2)
    if not ticks:
        plt.xticks([]), plt.yticks([])
        
def plot_4img(im1, tit1, im2, tit2, im3, tit3, im4, tit4, tam, ticks = False):
    plt.figure(figsize = tam)
    plt.subplot(1,4,1), plt.imshow(im1), plt.title(tit1)
    if not ticks:
        plt.xticks([]), plt.yticks([])
    plt.subplot(1,4,2), plt.imshow(im2), plt.title(tit2)
    if not ticks:
        plt.xticks([]), plt.yticks([])
    plt.subplot(1,4,3), plt.imshow(im3), plt.title(tit3)
    if not ticks:
        plt.xticks([]), plt.yticks([])
    plt.subplot(1,4,4), plt.imshow(im4), plt.title(tit4)
    if not ticks:
        plt.xticks([]), plt.yticks([])
    plt.show
```

```python
im_1 = cv2.imread(path + 'pinguim.jpg')
```

```python
im_1_ = cv2.cvtColor(im_1, cv2.COLOR_BGR2RGB)
```

```python
plot_2img(im_1_, 'Original', im_1, 'Original', (12,10))
```

```python
im_1_res1 = cv2.resize(im_1_, None, fx = .25, fy = .25, \
                      interpolation = cv2.INTER_CUBIC)
```

```python
im_1_res2 = cv2.resize(im_1_, None, fx = .10, fy = .10, \
                      interpolation = cv2.INTER_CUBIC)
```

```python
cv2.imwrite('im_1_original.jpg', im_1_)
cv2.imwrite('im_1_r25.jpg', im_1_res1)
cv2.imwrite('im_1_r10.jpg', im_1_res2)
```
